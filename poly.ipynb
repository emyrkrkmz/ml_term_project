```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
```

```python
X = pd.read_csv("dengue_features_train.csv")
y = pd.read_csv("dengue_labels_train.csv")

df = X.merge(y, on=["city","year","weekofyear"])

# Mevsimsellik (weekofyear -> sin/cos)
df["sin_week"] = np.sin(2*np.pi*df["weekofyear"]/52)
df["cos_week"] = np.cos(2*np.pi*df["weekofyear"]/52)
```

```python
DROP_COLS = ["city", "week_start_date", "weekofyear", "total_cases"]

FEATURES = [
    # zaman
    "year", "sin_week", "cos_week",
    # ndvi
    "ndvi_ne","ndvi_nw","ndvi_se","ndvi_sw",
    # yağış
    "precipitation_amt_mm","reanalysis_precip_amt_kg_per_m2",
    "reanalysis_sat_precip_amt_mm","station_precip_mm",
    # sıcaklık
    "reanalysis_air_temp_k","reanalysis_avg_temp_k",
    "reanalysis_min_air_temp_k","reanalysis_max_air_temp_k",
    "station_avg_temp_c","station_min_temp_c","station_max_temp_c",
    # nem/türev
    "reanalysis_relative_humidity_percent",
    "reanalysis_specific_humidity_g_per_kg",
    "reanalysis_dew_point_temp_k",
    "reanalysis_tdtr_k","station_diur_temp_rng_c",
]
```

```python
def time_split(city_df, val_frac=0.2):
    city_df = city_df.sort_values(["year","weekofyear"])
    n = len(city_df)
    cut = int((1 - val_frac) * n)
    train = city_df.iloc[:cut]
    val   = city_df.iloc[cut:]
    return train, val
```

```python
def make_pipeline(degree, alpha=1.0):
    return Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("ridge", Ridge(alpha=alpha))
    ])
```

```python
results = []

for city in ["sj", "iq"]:
    cdf = df[df["city"] == city].copy()
    train, val = time_split(cdf, val_frac=0.2)

    Xtr, ytr = train[FEATURES], train["total_cases"]
    Xva, yva = val[FEATURES],   val["total_cases"]

    for deg in [2,3,4]:
        pipe = make_pipeline(degree=deg, alpha=1.0)
        pipe.fit(Xtr, ytr)
        preds = pipe.predict(Xva)
        mae = mean_absolute_error(yva, preds)
        results.append({"city": city, "degree": deg, "MAE": mae})

pd.DataFrame(results).sort_values(["city","MAE"])
```

```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge

X_train = pd.read_csv("dengue_features_train.csv")
y_train = pd.read_csv("dengue_labels_train.csv")
X_test  = pd.read_csv("dengue_features_test.csv")
sub     = pd.read_csv("submission_format.csv")

def add_season(df):
    df = df.copy()
    df["sin_week"] = np.sin(2*np.pi*df["weekofyear"]/52)
    df["cos_week"] = np.cos(2*np.pi*df["weekofyear"]/52)
    return df

X_train = add_season(X_train)
X_test  = add_season(X_test)

train_df = X_train.merge(y_train, on=["city","year","weekofyear"])

FEATURES = [
    "year", "sin_week", "cos_week",
    "ndvi_ne","ndvi_nw","ndvi_se","ndvi_sw",
    "precipitation_amt_mm","reanalysis_precip_amt_kg_per_m2",
    "reanalysis_sat_precip_amt_mm","station_precip_mm",
    "reanalysis_air_temp_k","reanalysis_avg_temp_k",
    "reanalysis_min_air_temp_k","reanalysis_max_air_temp_k",
    "station_avg_temp_c","station_min_temp_c","station_max_temp_c",
    "reanalysis_relative_humidity_percent",
    "reanalysis_specific_humidity_g_per_kg",
    "reanalysis_dew_point_temp_k",
    "reanalysis_tdtr_k","station_diur_temp_rng_c",
]

BEST_DEGREE = {"iq": 2, "sj": 2}  # deg=2 best for both

def make_pipeline(degree, alpha=1.0):
    return Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("ridge", Ridge(alpha=alpha)),
    ])

preds_all = []

for city in ["sj","iq"]:
    deg = BEST_DEGREE[city]
    pipe = make_pipeline(degree=deg)

    tr = train_df[train_df["city"] == city].sort_values(["year","weekofyear"])
    te = X_test[X_test["city"] == city].sort_values(["year","weekofyear"])

    pipe.fit(tr[FEATURES], tr["total_cases"])
    preds = pipe.predict(te[FEATURES])

    out = te[["city","year","weekofyear"]].copy()
    out["total_cases"] = preds
    preds_all.append(out)

pred_df = pd.concat(preds_all, ignore_index=True)

sub = sub.drop(columns=["total_cases"]).merge(pred_df, on=["city","year","weekofyear"], how="left")

sub["total_cases"] = sub["total_cases"].clip(lower=0).round().astype(int)

sub.to_csv("submission_model1_poly.csv", index=False)
sub.head(), sub.isna().sum()
```
